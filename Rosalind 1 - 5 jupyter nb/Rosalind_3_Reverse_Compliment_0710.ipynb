```
complements = { 'A' : 'T', 'T' : 'A', 'C' : 'G', 'G' : 'C'}
```

```
result = ''
target = (str(input("Yeet: ")))
target = target[::-1]
for ch in target:
  try:
    result += complements.get(ch)
  except:
    print("yeet?")
    break
    
print (result)
```
